```python
n = len(df["Reviewer's id"])
for i in range(n):
    if df["Reviewer's rate"][i] == 4 or df["Reviewer's rate"][i] == 5:
        df['label'][i] = "Positive"
    else:
        if df["Reviewer's rate"][i] == 3:
            df['label'][i] = "neutral"
        else:
             if df["Reviewer's rate"][i] == 2 or df["Reviewer's rate"][i] == 1:
                df['label'][i] = "negative"
    if i == n:
      print('đã hoàn thành')
```
