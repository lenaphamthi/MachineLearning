**Đánh giá mô hình hồi quy tuyến tính đơn giản trên tập huấn luyện**

```python
def danh_gia_mo_hinh(y_test, y_hat):
  #(i) accuracy
  i = accuracy_score(y_test,y_hat)
  #(ii) precision
  # Tính độ chính xác của từng nhãn lớp
  ii_None = precision_score(y_test,y_hat , average=None)
  # Tính tổng số lượng dữ liệu được dự đoán đúng so với tập dữ liệu
  ii_micro = precision_score(y_test,y_hat , average='micro')
  #(iii) recall
  iii = recall_score(y_test,y_hat, average = None)
  #(iv) f1-measure
  iv = list(f1_score(y_test, y_hat, average=None))
  return i, ii_None, iii, iv
```

* Ưu điểm của SVM

Tính chính xác cao: SVM thường có độ chính xác cao hơn LR, đặc biệt là đối với các dữ liệu có sự phân tách tuyến tính.
Độ bền: SVM ít bị ảnh hưởng bởi nhiễu dữ liệu hơn LR.
Tính linh hoạt: SVM có thể được sử dụng cho nhiều loại bài toán phân loại, bao gồm phân loại tuyến tính và phi tuyến.

* Nhược điểm của SVM

Yêu cầu nhiều bộ dữ liệu: SVM thường cần nhiều bộ dữ liệu hơn LR để đạt được độ chính xác cao.
Khó giải thích: SVM có thể khó giải thích hơn LR, vì nó không tạo ra một mô hình tuyến tính có thể dễ dàng hiểu được.


---


* Ưu điểm của LR

Tốc độ học nhanh: LR có thể học nhanh hơn SVM, đặc biệt là đối với các bộ dữ liệu lớn.
Dễ giải thích: LR tạo ra một mô hình tuyến tính có thể dễ dàng hiểu được.
Thích ứng với các dữ liệu không cân bằng: LR có thể thích ứng tốt hơn với các dữ liệu không cân bằng hơn SVM.
* Nhược điểm của LR

Độ chính xác thấp hơn SVM: LR thường có độ chính xác thấp hơn SVM, đặc biệt là đối với các dữ liệu có sự phân tách tuyến tính.
Dễ bị ảnh hưởng bởi nhiễu dữ liệu: LR có thể bị ảnh hưởng bởi nhiễu dữ liệu hơn SVM.


---


* So sánh hiệu suất của SVM và LR

Nhìn chung, SVM có độ chính xác cao hơn LR, đặc biệt là đối với các dữ liệu có sự phân tách tuyến tính. Tuy nhiên, SVM yêu cầu nhiều bộ dữ liệu hơn LR và có thể khó giải thích hơn. LR có tốc độ học nhanh hơn SVM, dễ giải thích hơn và có thể thích ứng tốt hơn với các dữ liệu không cân bằng.

* Chọn thuật toán nào?

Việc lựa chọn thuật toán SVM hay LR phụ thuộc vào nhiều yếu tố, bao gồm dữ liệu đầu vào, mục tiêu phân loại và các yêu cầu cụ thể của ứng dụng. Nếu dữ liệu có sự phân tách tuyến tính và yêu cầu độ chính xác cao, thì SVM là một lựa chọn tốt. Nếu yêu cầu tốc độ học nhanh, khả năng giải thích và khả năng thích ứng với các dữ liệu không cân bằng, thì LR là một lựa chọn tốt hơn.
